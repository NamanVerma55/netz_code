```python
import pdfplumber
import pandas as pd
import os
```

```python
def extract_tables_to_csv(pdf_path, output_folder):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    
    with pdfplumber.open(pdf_path) as pdf:
        for page_number, page in enumerate(pdf.pages, start=1):
            # Extract tables from the current page
            tables = page.extract_tables()
            
            for table_number, table in enumerate(tables, start=1):
                if table:  # Check if the table is not empty
                    # Convert table to a DataFrame
                    df = pd.DataFrame(table[1:], columns=table[0])  # First row as column headers
                    
                    # Generate output file name
                    output_file = f"{output_folder}/page_{page_number}_table_{table_number}.csv"
                    
                    # Save DataFrame to CSV
                    df.to_csv(output_file, index=False)
                    print(f"Saved table from page {page_number}, table {table_number} to {output_file}")

# Path to the PDF file
pdf_path = r'D:\project\netz_work\OCR\SIW Issue 363 09_04_1999 1.pdf'

# Output folder for CSV files
output_folder = r'D:\project\netz_work\OCR\output_csv'

# Run the extraction function
extract_tables_to_csv(pdf_path, output_folder)
```
